Nesse notebook, analisamos a partir do dataframe clássico do seaborn Iris a influência da normalização dos dados na acurácia do algoritmo knn.

```python
import seaborn as sns
```

Primeiramente, carregamos o dataframe e o visualizamos.

```python
df = sns.load_dataset('iris')
RMSEl, mael = [], []
df
```

Usaremos sepal_length, sepal_width e petal_width para prever petal_length.
Primeiro, selecionamos os dados para treino e teste:

```python
from sklearn.model_selection import train_test_split
TAMANHO_TESTE = 0.1
SEMENTE_ALEATORIA = 27

indices = df.index
indices_treino, indices_teste = train_test_split(
    indices, test_size=TAMANHO_TESTE, random_state=SEMENTE_ALEATORIA
)

df_treino = df.loc[indices_treino]
df_teste = df.loc[indices_teste]

ATRIBUTOS = ['sepal_length', 'sepal_width', 'petal_width']
TARGET = ['petal_length']
X_treino = df_treino.reindex(ATRIBUTOS, axis=1)
y_treino = df_treino.reindex(TARGET, axis=1)

X_teste = df_teste.reindex(ATRIBUTOS, axis=1)
y_teste = df_teste.reindex(TARGET, axis=1)
```

Agora, treinamos o modelo sem normalizar os dados:

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error

NUM_VIZINHOS = 3
modelo_knn = KNeighborsRegressor(n_neighbors=NUM_VIZINHOS)

modelo_knn.fit(X_treino, y_treino)

from sklearn.metrics import root_mean_squared_error

X_verdadeiro = X_teste
y_verdadeiro = y_teste

y_previsto = modelo_knn.predict(X_verdadeiro)

RMSE = root_mean_squared_error(y_verdadeiro, y_previsto)
mae = mean_absolute_error(y_verdadeiro, y_previsto)

RMSEl.append(RMSE)
mael.append(mae)

print(f"O RMSE do modelo foi de {RMSE} unidades de y.")
print(f"Erro médio absoluto: {mae:.2f} pontos")
```

Agora, normalizemos usando StandarScaller:

```python
from sklearn.preprocessing import StandardScaler

df_ss = df.copy()
normalizador = StandardScaler()

x= df["petal_width"].values.reshape(-1,1)
normalizador.fit(x)
pw_norm = normalizador.transform(x)
df_ss['petal_width'] = pw_norm

x= df["sepal_width"].values.reshape(-1,1)
normalizador.fit(x)
sw_norm = normalizador.transform(x)
df_ss['sepal_width'] = sw_norm

x= df["sepal_length"].values.reshape(-1,1)
normalizador.fit(x)
sl_norm = normalizador.transform(x)
df_ss['sepal_length'] = sl_norm
```

```python
df_ss
```

```python
indices = df_ss.index
indices_treino, indices_teste = train_test_split(
    indices, test_size=TAMANHO_TESTE, random_state=SEMENTE_ALEATORIA
)

df_treino = df_ss.loc[indices_treino]
df_teste = df_ss.loc[indices_teste]

ATRIBUTOS = ['sepal_length', 'sepal_width', 'petal_width']
TARGET = ['petal_length']
X_treino = df_treino.reindex(ATRIBUTOS, axis=1)
y_treino = df_treino.reindex(TARGET, axis=1)

X_teste = df_teste.reindex(ATRIBUTOS, axis=1)
y_teste = df_teste.reindex(TARGET, axis=1)
```

```python
modelo_knn.fit(X_treino, y_treino)

X_verdadeiro = X_teste
y_verdadeiro = y_teste

y_previsto = modelo_knn.predict(X_verdadeiro)

RMSE = root_mean_squared_error(y_verdadeiro, y_previsto)
mae = mean_absolute_error(y_verdadeiro, y_previsto)

RMSEl.append(RMSE)
mael.append(mae)

print(f"O RMSE do modelo foi de {RMSE} unidades de y.")
print(f"Erro médio absoluto: {mae:.2f} pontos")
```

Enfim, feito todos os modelos, analisemos cada um:

```python
import pandas as pd
d = {
    'norm': ['Nd', 'Std', 'MaxAbs', 'MinMax'],
    'RMSE': RMSEl,
    'mae': mael}
df_f = pd.DataFrame(d)
df_f = df_f.T
df_f
```

Primeiramente, é perceptível que o standard e maxabs tiveram desempenhos muito similares. Além disso, vemos que algo não esperado aconteceu, não normalizar teve desenpenho superior aos dois comentados. Porém o MinMax teve um desenpenho considerávelmente maior em comparação aos outros.
Os dois primeiros normalizadores terem diminuido desempenho pode ter ocorrido por azar nos dados utilizados, porém, é um indicativo de que o MinMax pode ser o melhor normalizador para o algoritmo knn. Mais testes com novos dataframes devem ser feitos, para confirmarmos ou negarmos as teorias elaboradas.

### Conclusão
O MinMax obteve maior desempenho em comparação aos demais normalizadores, sendo provável de ser o mais indicado para usar no algoritmo knn. Além disso, vemos como normalizar impactou em todos os casos o resultado previsto.
